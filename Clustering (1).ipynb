Hierarchical Clustering

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
min_max_scaler = MinMaxScaler()
norm = min_max_scaler.fit_transform(EastWestAirlines)
norm
```

```python
len(norm)
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
plt.figure(figsize = (17,11))
dendrogram=sch.dendrogram(sch.linkage(norm, method='complete'))
```

```python

# Agglomerative Clustering
hc = AgglomerativeClustering(n_clusters=5, linkage = 'complete')
```

```python

y_hc = hc.fit_predict(norm)
y_hc
```

```python
hc.labels_
```

```python
EastWestAirlines['h_clusterid'] = hc.labels_
```

K-means Clustering

```python
Data=EastWestAirlines.drop(columns='h_clusterid')
Data.head()
```

```python
from sklearn.cluster import KMeans
scaler = StandardScaler().fit_transform(Data)
scaler
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaler)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
wcss
```

```python
clusters = KMeans(6, random_state=42)
clusters.fit(scaler)
clusters.labels_
```

```python
clusters.fit_predict(scaler)
```

```python
clusters.inertia_
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(Data, clusters.labels_)
```

```python
c2=KMeans(2, random_state=41)
c2.fit(scaler)
```

```python
silhouette_score(Data, c2.labels_)
```

```python
Data['clusterid']=clusters.labels_
Data
```

```python
Data['clusterid'].value_counts()
```

```python
Data.groupby('clusterid').mean()
```
